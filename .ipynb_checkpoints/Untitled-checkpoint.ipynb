```python
primary['total_votes'] = primary[['Clinton','Trump']].sum(axis=1)
primary['Clinton'] = primary['Clinton']/primary['total_votes']
primary['Trump']   = primary['Trump']/primary['total_votes']
```

```python
# aggregate states
nyt_results_state = primary[['Trump','State']].groupby('State').mean()
nyt_results_state.head()
```

```python
# geo_edu = pd.merge(geodf, census_state, left_index=True, right_index=True)
allData = pd.merge(demographic, primary, how="inner", left_on = 'state_abbreviation',right_on = 'state_abbreviation')
# allData
del allData['Vote by town']
del allData['Vote by county']
allData
# allData = allData[['Clinton','Trump','State','Vote by county','Place']]
```

```python
for col in allData.columns:
    print(col)
```

```python
col = u'Bachelors'
print(allData[col].min(), allData[col].max())
print (allData[col])
allData[col].hist(figsize=[16,4])
```

```python
col = u'White'
print(allData[col].min(), allData[col].max())
print (allData[col])
allData[col].hist(figsize=[16,4])
```

```python
allData['color'] = allData[col].apply(np.log10)

cmap = 'viridis'
vmin = allData['color'].min()
vmax = allData['color'].max()
cbar_vmin = allData[col].min()
cbar_vmax = allData[col].max()

print 'vmin, vmax = ',vmin, vmax
print 'cbar_vmin, cbar_vmax = ',cbar_vmin, cbar_vmax

fig, ax = plt.subplots(3,1,figsize=[8,8], sharex=True, sharey=True)

g = allData.plot(column='color',cmap=cmap, edgecolor='white',
                 linewidth = 0.5,vmin=vmin, vmax=vmax, ax=ax[0], alpha=1)
g.set_xlim([-130,-65])
g.set_ylim([24,50])
# add colorbar
axbb = np.array(ax[0].get_position()).flatten()
shift = 0.025
cbar_width = 0.05
cax = fig.add_axes([axbb[2]+shift,axbb[1],cbar_width,axbb[3]-axbb[1]])
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=cbar_vmin, vmax=cbar_vmax))
# fake up the array of the scalar mappable. Urgh...
sm._A = []
fig.colorbar(sm, cax=cax)
ax[0].set_title(col.replace('_', ' '), fontsize=16)

g = allData.plot(column='Trump',cmap='coolwarm', edgecolor='white',linewidth = 0.5, ax=ax[1], alpha=1)
# add colorbar
axbb = np.array(ax[1].get_position()).flatten()
shift = 0.025
cbar_width = 0.05
cax = fig.add_axes([axbb[2]+shift,axbb[1],cbar_width,axbb[3]-axbb[1]])
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=plt.Normalize(vmin=allData['Trump'].min(), vmax=allData['Trump'].max()))
# fake up the array of the scalar mappable. Urgh...
sm._A = []
fig.colorbar(sm, cax=cax)
ax[1].set_title('Percentage for Trump', fontsize=16)
```
